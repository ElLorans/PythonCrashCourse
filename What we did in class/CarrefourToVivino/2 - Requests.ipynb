# requests tutorial
Requests is a library for http requests (e.g.: going to a website)

```python
import requests
```

```python
url = "https://en.wikipedia.org/wiki/Chianti"
response = requests.get(url)
```

### 200 is a standard http response for OK
See https://developer.mozilla.org/en-US/docs/Web/HTTP/Status

```python
print(response)
```

### response.text is what your browser "reads" when going to https://en.wikipedia.org/wiki/Chianti
You can check by going to https://en.wikipedia.org/wiki/Chianti and pressing Ctrl+u on Windows or Cmd+u on OS

```python
print(response.text)
```
